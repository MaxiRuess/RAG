```python
DIR = os.path.normpath(
    os.path.join(os.path.dirname(os.path.abspath("/Users/maximilianruess/Documents/GitHub/RAG/notebook.ipynb")), "..", "utils")
)
```

```python
class Model: 
    def __init__(self, model_id: str = "google/gemma-2b-it", device = 'mps'): 
        
        ACCESS_TOKEN = os.getenv("ACCESS_TOKEN")
        
        self.tokenizer = AutoTokenizer.from_pretrained(model_id, 
                                                       cache_dir=DIR)
        
        self.model = AutoModelForSequenceClassification.from_pretrained(model_id, 
                                                                       cache_dir=DIR, 
                                                                       device_map = "auto", 
                                                                       token = ACCESS_TOKEN)
        
        self.model = torch.quantization.quantize_dynamic(
            self.model, {torch.nn.Linear}, dtype=torch.qint8)
        
        
        self.model.eval()
        self.chat = []
        self.device = device 
        
model = Model()
```
